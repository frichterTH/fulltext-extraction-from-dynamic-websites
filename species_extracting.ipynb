```python
import pandas as pd
import spacy
import re
```

```python
df = pd.read_csv("fulltext_articles.csv")
list_text = df["title"].dropna().tolist()
text = ' '.join(map(str, list_text))
```

```python
nlp = spacy.load('en_core_sci_lg')
```

```python
doc_list = []
for i in list_text[:20]:
    doc = nlp(i)
    doc_list.append(doc)
```

```python
for i in doc_list[6]:
    print(i, i.pos_)
```

```python
# get all bacteria species with a certain pattern from titles
# findall in case there are multiple species in the title
species =[]
for i in list_text:
    match = re.findall('^\w+\s\w+\s\w+\.\s\w+\.', i) 
    species.append(match)
    
```

```python
species
```

```python
# flattens nested list
species_flattened = sum(species, [])
```

```python
len(species)
```

```python
len(species_flattened)
```

```python
# removes duplicates
species_list = list(dict.fromkeys(species_flattened))
```

```python
len(species_list)
```

```python
species_list
```

```python
# remove non-bacterial string
species_list.remove('CANDIDATUS LIST No. 3.')
```

```python
# splits species names and keeps first two words 
species_split = []
for name in species_list:
    name_split = (name.split()[0:2])
    species_split.append(' '.join(name_split))
    
```
